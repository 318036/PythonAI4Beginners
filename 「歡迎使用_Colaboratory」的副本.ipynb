```
# 📦 安裝必要套件（Colab 預設已安裝 jieba 和 bs4）
!pip install lxml

# ✅ 載入必要套件
import jieba
import jieba.analyse
from bs4 import BeautifulSoup
from collections import Counter
from google.colab import files

# 🔕 關閉 jieba 除錯訊息
jieba.setLogLevel(jieba.logging.INFO)

# ⬆️ 上傳 HTML 檔案
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# 📄 讀取 HTML 內容
with open(file_name, "r", encoding="utf-8") as file:
    html_content = file.read()

# 🌐 解析 HTML 並提取 <p class="card-content"> 文字
soup = BeautifulSoup(html_content, "lxml")
paragraphs = soup.find_all("p", class_="card-content")
text_content = " ".join(p.get_text().strip() for p in paragraphs)

# 🛑 停用詞（可自訂）
stop_words = set([
    '所以', '好', '因為', '大家', '的', '是', '了', '我', '也', '在', '和', '就',
    '不', '有', '他', '她', '你', '我們', '這個',
    'https', 'colab', 'com', 'research', 'google', 'drive', 'usp', 'sharing', 'scrollTo'
])

# ✂️ 分詞 + 過濾停用詞
words = jieba.cut(text_content)
filtered_words = [word for word in words if word not in stop_words]

# 📊 統計詞頻
word_count = Counter(filtered_words)
filtered_word_count = {word: count for word, count in word_count.items() if count > 1}

# 🔑 擷取關鍵詞（TF-IDF）
keywords = jieba.analyse.extract_tags(" ".join(filtered_words), topK=15, withWeight=True)

# 🔍 詞頻查詢
user_input = input("請輸入要查詢的詞：")
print(f"\n「{user_input}」出現了 {word_count.get(user_input, 0)} 次。")

# 📢 輸出高頻詞
print("\n🔥 高頻詞（出現次數 > 1）:")
for word, count in sorted(filtered_word_count.items(), key=lambda x: x[1], reverse=True)[:15]:
    print(f"{word}: {count}")

# 📌 輸出關鍵詞（TF-IDF）
print("\n🔑 文章關鍵詞（TF-IDF 權重）:")
for kw, weight in keywords:
    print(f"{kw}: {weight:.4f}")
```
