```
# ğŸ“¦ å®‰è£å¿…è¦å¥—ä»¶ï¼ˆColab é è¨­å·²å®‰è£ jieba å’Œ bs4ï¼‰
!pip install lxml

# âœ… è¼‰å…¥å¿…è¦å¥—ä»¶
import jieba
import jieba.analyse
from bs4 import BeautifulSoup
from collections import Counter
from google.colab import files

# ğŸ”• é—œé–‰ jieba é™¤éŒ¯è¨Šæ¯
jieba.setLogLevel(jieba.logging.INFO)

# â¬†ï¸ ä¸Šå‚³ HTML æª”æ¡ˆ
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# ğŸ“„ è®€å– HTML å…§å®¹
with open(file_name, "r", encoding="utf-8") as file:
    html_content = file.read()

# ğŸŒ è§£æ HTML ä¸¦æå– <p class="card-content"> æ–‡å­—
soup = BeautifulSoup(html_content, "lxml")
paragraphs = soup.find_all("p", class_="card-content")
text_content = " ".join(p.get_text().strip() for p in paragraphs)

# ğŸ›‘ åœç”¨è©ï¼ˆå¯è‡ªè¨‚ï¼‰
stop_words = set([
    'æ‰€ä»¥', 'å¥½', 'å› ç‚º', 'å¤§å®¶', 'çš„', 'æ˜¯', 'äº†', 'æˆ‘', 'ä¹Ÿ', 'åœ¨', 'å’Œ', 'å°±',
    'ä¸', 'æœ‰', 'ä»–', 'å¥¹', 'ä½ ', 'æˆ‘å€‘', 'é€™å€‹',
    'https', 'colab', 'com', 'research', 'google', 'drive', 'usp', 'sharing', 'scrollTo'
])

# âœ‚ï¸ åˆ†è© + éæ¿¾åœç”¨è©
words = jieba.cut(text_content)
filtered_words = [word for word in words if word not in stop_words]

# ğŸ“Š çµ±è¨ˆè©é »
word_count = Counter(filtered_words)
filtered_word_count = {word: count for word, count in word_count.items() if count > 1}

# ğŸ”‘ æ“·å–é—œéµè©ï¼ˆTF-IDFï¼‰
keywords = jieba.analyse.extract_tags(" ".join(filtered_words), topK=15, withWeight=True)

# ğŸ” è©é »æŸ¥è©¢
user_input = input("è«‹è¼¸å…¥è¦æŸ¥è©¢çš„è©ï¼š")
print(f"\nã€Œ{user_input}ã€å‡ºç¾äº† {word_count.get(user_input, 0)} æ¬¡ã€‚")

# ğŸ“¢ è¼¸å‡ºé«˜é »è©
print("\nğŸ”¥ é«˜é »è©ï¼ˆå‡ºç¾æ¬¡æ•¸ > 1ï¼‰:")
for word, count in sorted(filtered_word_count.items(), key=lambda x: x[1], reverse=True)[:15]:
    print(f"{word}: {count}")

# ğŸ“Œ è¼¸å‡ºé—œéµè©ï¼ˆTF-IDFï¼‰
print("\nğŸ”‘ æ–‡ç« é—œéµè©ï¼ˆTF-IDF æ¬Šé‡ï¼‰:")
for kw, weight in keywords:
    print(f"{kw}: {weight:.4f}")
```
